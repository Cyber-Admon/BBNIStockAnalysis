### Exploring the Data Set

```python
data = read_csv("BBNI.JK_weekly.csv")
```

```python
data["Volatility"] = data["Close"].std()
```

```python
data["Volatility"].sort_values(ascending=False)
```

```python
data.head()
```

```python
data["Volatility"].plot(kind="line")
```

```python
data["Volatility"].plot(kind="line")
```

```python
import matplotlib.pyplot as plt

# Specify figure size and resolution in a single call
plt.figure(figsize=(100, 15), dpi=100)

# Create the line plot
plt.plot(data["Date"], data["Close"], linewidth=3)

# Add labels
plt.xlabel("Date")
plt.ylabel("Price")
plt.title("Stock Price Data Analysis") 
# Iterate over data and annotate with concise syntax
for date, row in data.iterrows():
    plt.annotate("High", xy=(date, row["High"]), fontsize=5)
    plt.annotate("Low", xy=(date, row["Low"]), fontsize=5)

# Adjust layout for better readability
plt.tight_layout()

# Show the plot
plt.show()
```
